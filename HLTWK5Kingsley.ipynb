```python
interval = alt.selection_interval()

alt.Chart(df).mark_bar().encode(
    x='Positive_Review',
    y='Reviewer_Score',
    color='Reviewer_Nationality'
    ).properties(
        selection=interval
    )
```

```python
interval = alt.selection_interval()

alt.Chart(df).mark_bar().encode(
    x='Positive_Review',
    y='Reviewer_Nationality',
    color=alt.condition(interval,'Reviewer_Nationality',alt.value('lightgray'))
    ).properties(
        selection=interval
    )
```

```python
interval = alt.selection_interval()

alt.Chart(df).mark_point().encode(
    x='Reviewer_Nationality',
    y='Positive_Review',
    color=alt.condition(interval,'Review_Date',alt.value('lightgray'))
    ).properties(
        selection=interval
    )
```

```python
interval = alt.selection_interval()

scatter = alt.Chart(df).mark_point().encode(
    x='Reviewer_Nationality',
    y='Positive_Review',
    color=alt.condition( interval,'Review_Date',alt.value('lightgray'))
    ).properties(
        selection=interval
    )
bar = alt.Chart(df).mark_bar().encode(
    x='Reviewer_Nationality',
    y= 'average(Positive_Review)',
    color='Review_Date'
).transform_filter(interval
)

alt.vconcat(scatter, bar)
```
